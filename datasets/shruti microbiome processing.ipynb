```python
%%bash
FILE="/content/File_Storage/PRJEB4927_Download_Script.txt"

if [[ ! -f "$FILE" ]]; then
  echo "Error: File '$FILE' not found."
  exit 1
fi

# Replace 'wget -nc' with 'curl -O' in-place
sed -i 's/wget\s\+-nc/curl -O/g' "$FILE"

echo "Replaced 'wget -nc' with 'curl -O' in $FILE."

chmod +x "/content/File_Storage/PRJEB4927_Download_Script.txt"
cd /content/File_Storage
./PRJEB4927_Download_Script.txt
```

Now, we have downloaded all of the raw .fastq files. From here on, we will create output files, also known as artifacts. We will be utilizing these to then process our data into a table.

```python
%%bash
echo -e "sample-id\tabsolute-filepath" >> /content/Output_Files/humanGenomeInformation.tsv

for file in /content/File_Storage/*.fastq.gz; do
  filename=$(basename "$file")                         # e.g. ERR1513668.fastq.gz
  sample_id="${filename%.fastq.gz}"                    # strips .fastq.gz → ERR1513668
  echo -e "${sample_id}\t${file}" >> /content/Output_Files/humanGenomeInformation.tsv
done
```

```python
%%bash
qiime tools import \
  --type 'SampleData[SequencesWithQuality]' \
  --input-path /content/Output_Files/humanGenomeInformation.tsv \
  --output-path /content/Output_Files/demultiplexedOutput.qza \
  --input-format SingleEndFastqManifestPhred33V2
```

```python
%%bash
qiime tools import \
  --type 'FeatureData[Sequence]' \
  --input-path /content/be2e4caded84dcdcba8b2b0c0e0771d3/gg_13_8_otus/rep_set/94_otus.fasta \
  --output-path /content/Output_Files/94_otus.qza
```

```python
%%bash
qiime tools import \
  --type 'FeatureData[Taxonomy]' \
  --input-format HeaderlessTSVTaxonomyFormat \
  --input-path /content/be2e4caded84dcdcba8b2b0c0e0771d3/gg_13_8_otus/taxonomy/94_otu_taxonomy.txt \
  --output-path /content/Output_Files/94_otu_taxonomy.qza
```

By now you should have your 94_otus.qza and your 94_otu_taxonomy.gza. Note, here we are assuming 94% similarity as the similarity threshold. If you prefer having another value as the threshold instead, you could use that specific _value._otus.fasta file from the rep_set folder and the specific _value_otu_taxonomy.txt file from that specific taxonomy folder. Both the rep_set and taxonomy folder is inside here: /content/be2e4caded84dcdcba8b2b0c0e0771d3/gg_13_8_otus

Below, we will go ahead and create our own customer feature classifier. The script below may take up to 40 minutes for it to load successfully to create the 94_otus_customer_classifier.qza

```python
%%bash
qiime feature-classifier fit-classifier-naive-bayes \
  --i-reference-reads /content/Output_Files/94_otus.qza \
  --i-reference-taxonomy /content/Output_Files/94_otu_taxonomy.qza \
  --o-classifier /content/Output_Files/94_otus_customer_classifier.qza
```

```python
%%bash
qiime feature-classifier classify-sklearn \
  --i-classifier /content/Output_Files/94_otus_customer_classifier.qza \
  --i-reads /content/File_Storage/dada2/representative_sequences.qza \
  --o-classification /content/Output_Files/94_classification_taxonomy.qza
```

Before being able to generate the OTU table with the taxonomic labels, we need to ensure that our tables discard any of the ASV so that both the table.qza and taxonomy.qza are matching. After this

```python
%%bash
qiime taxa collapse \
  --i-table /content/File_Storage/dada2/table.qza \
  --i-taxonomy /content/Output_Files/94_classification_taxonomy.qza \
  --p-level 6 \
  --o-collapsed-table /content/Output_Files/94_OTUtable.qza
```

```python
import pandas as pd

url = "https://www.ebi.ac.uk/ena/portal/api/filereport?accession=ERP016332&result=read_run&fields=run_accession,sample_alias,age,sex,disease,status&format=tsv"
df = pd.read_csv(url, sep="\t")
df.to_csv('ena_metadata.csv', index=False)
```

The place to obtain the metadata is here: https://qiita.ucsd.edu/study/description/10567. Every project or study will have a metadata file that you need to find. In this particular case, the study is hosted by QIITA so the metadata was found there. The metadata file will vary from project to project.

To obtain the metadata file, you will need to log in/ create an account. The study is numbered as 10567 so you can go into that specific study. Once entering that study/ project, you can click "Sample Information" that you can download from the top left. I downloaded the file and without any processing, the text file has been added and it is labelled as qitta_metadata.txt

```python
qitta_metadata = pd.read_csv('/content/Metadata_Downloaded/qitta_metadata.txt', sep='\t')  # assuming tab-delimited

qitta_metadata.to_csv('/content/Output_Files/qitta_metadata.csv', index=False)

ena_metadata = pd.read_csv('/content/File_Storage/ena_metadata.csv')
ena_metadata['sample_alias'] = ena_metadata['sample_alias'].str.split(':').str[-1]
ena_metadata.to_csv('/content/Output_Files/ena_metadata.csv', index=False)

metadata = pd.merge(qitta_metadata, ena_metadata, right_on='sample_alias', left_on='sample_name')
metadata = metadata.rename(columns={'run_accession': 'sample-id'})
metadata = metadata.rename(columns={'sex_x': 'sex'})
metadata = metadata.rename(columns={'age_x': 'age'})



metadata.to_csv('metadata.csv', index=False)

target_columns = ['sample_name', 'sample-id', 'sex', 'age', 'parkinson_disease']
processed_metadata = metadata[target_columns]
processed_metadata.to_csv('/content/Output_Files/processed_metadata.csv', index=False)
processed_metadata.head()
```

```python
otu_table = pd.read_csv('/content/Output_Files/94_featureOTUtable/feature-table.tsv', sep="\t", skiprows=1, index_col=0).T
otu_table = otu_table.reset_index()
otu_table = otu_table.rename(columns={'index': 'OTU_ID'})
otu_table['Total_reads'] = otu_table.drop(columns=['OTU_ID']).sum(axis=1)
processed_metadata = pd.read_csv('/content/Output_Files/processed_metadata.csv')

final_table = pd.merge(processed_metadata, otu_table, left_on='sample-id', right_on='OTU_ID')
final_table.to_csv('/content/Output_Files/final_table.csv', index=False)
final_table.head()
```
